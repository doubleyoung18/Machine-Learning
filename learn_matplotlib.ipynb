```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1, 1, 50)
y = 2 * x + 1

plt.figure()  # 定义一个图像窗口
plt.plot(x, y)  # 画曲线
plt.show()  # 显示图像
```

```python
x = np.linspace(-3, 3, 50)
y1 = x ** 2
y2 = 2 * x + 1

plt.figure(num=2, figsize=(8, 5))  # 设置窗口编号、大小
plt.plot(x, y1)
plt.plot(x, y2, color='red', linewidth=1.0, linestyle='--')  # 设置曲线样式
plt.show()
```

```python
# 画出基本曲线
x = np.linspace(-3, 3, 50)
y = 2 * x + 1
plt.figure(num=5, figsize=(10, 6))
plt.plot(x, y)
# 移动坐标轴
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# ax.spines['bottom'].set_color('none')
# ax.spines['left'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
x0 = 1
y0 = x0 * 2 + 1
plt.plot([x0, x0], [0, y0], 'k--', linewidth=2.0)  # 画一条垂直于x轴的虚线
plt.scatter([x0, ], [y0], s=50, color='b')  # 设置点样式
# 添加注释annotate
plt.annotate(r'$2x+1=%s$' % y0,  # 内容
             xy=(x0, y0),  # 指向
             xycoords='data',  # 基于数值选择位置
             xytext=(+30, -30),  # 描述标注位置 
             textcoords='offset points',  # xy偏差值 
             fontsize=16, 
             arrowprops=dict(arrowstyle='->', 
                             connectionstyle='arc3, rad=.2')  # 设置箭头类型
            )
# 添加注释text
plt.text(-3.7, 3,  # 选取text位置
         r'$This\ is\ the\ some\ text. \mu\ \sigma_i\ \alpha_t$',  # 内容 
         fontdict={'size' : 16, 'color' : 'r'}  # 设置文本字体
        )
plt.show()
```

```python
x = np.linspace(-3, 3, 50)
y = x * 0.1
plt.figure()
# 在plt v2.02+中，设置zorder给plot在z轴方向排序
plt.plot(x, y, linewidth=10, zorder=1)
plt.ylim(-2, 2)
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
# 调整坐标
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)  # 调整字体大小
    label.set_bbox(dict(facecolor='white',  # 设置前景色 
                        edgecolor='None',   # 设置边框
                        alpha=0.7,  # 设置透明度
                        zorder=2))  # z轴方向排序
plt.show()
```

```python
# 散点图
n = 1024
X = np.random.normal(0, 1, n)  # 设置每一个点的X值
Y = np.random.normal(0, 1, n)  # 设置每一个点的y值
T = np.arctan2(Y, X)
plt.scatter(X, Y, s=74, c=T, alpha=.5)
plt.xlim(-1.5, 1.5)
plt.xticks(())  # 隐藏x轴
plt.ylim(-1.5, 1.5)
plt.yticks(())  # 隐藏y轴
plt.show()
```

```python
# 柱状图
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y1)
plt.bar(X, -Y2)

plt.xlim(-.5, n)
plt.xticks(())
plt.ylim(-1.25, 1.25)
plt.yticks(())
# 增加颜色
plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')  # facecolor 主题色
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')  # edgecolor 边框色
# 增加数据
# ha:horizontal alignment 横向对齐
# va:vertical alignment 纵向对齐
for x, y in zip(X, Y1):
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va='bottom')
for x, y in zip(X, Y2):
    plt.text(x + 0.4, -y - 0.05, '%.2f' % y, ha='center', va='top')
plt.show()
```

```python
# 等高线
def f(x,y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X ,Y = np.meshgrid(x, y)
# 颜色填充
plt.contourf(X, Y, f(X, Y), 8, alpha=.5, cmap=plt.cm.hot)
# 绘制等高线
C = plt.contour(X, Y, f(X, Y), 9, colors='black')
# num表示等高线密集程度，共0-9是个等级
# 添加高度数字
plt.clabel(C, inline=True, fontsize=10)  # inline默认为True
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
# 随机矩阵画图
a = np.array([0.313660827978, 0.365348418405, 0.423733120134,
              0.365348418405, 0.439599930621, 0.525083754405,
              0.423733120134, 0.525083754405, 0.651536351379]
            ).reshape(3,3)
plt.imshow(a, interpolation='nearest',   # 出图方式(内插法)
           # None, none, nearest, bilinear, bicubic, spline16,
           # spline36, hanning, hamming, hermite, kasier, quadric,
           # catrom, gaussian, bessel, mitchell, sinc, lanczos
           cmap='bone', 
           origin='lower')  # 选择原点位置
plt.colorbar(shrink=.92)  # 添加颜色标尺
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
# 3D图
from mpl_toolkits.mplot3d import Axes3D  # 3D坐标轴显示模块
fig = plt.figure()  # 设置一个图像窗口
ax = Axes3D(fig)  # 在窗口上添加3D坐标轴
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X ** 2 + Y ** 2)
Z = np.sin(R)
ax.plot_surface(X, Y, Z, rstride=1,  # rstride行跨度
                cstride=1,  # cstride列跨度
                cmap=plt.get_cmap('rainbow'))
# 投影
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.get_cmap('rainbow'))
# 添加XY平面的等高线
ax.contourf(X, Y, Z, zdir='x', offset=-5, cmap=plt.get_cmap('rainbow'))
# 添加YZ平面的等高线
ax.set_zlim(-2, 2)
plt.show()
```

```python
# Subplot分格显示
plt.figure()
ax1 = plt.subplot2grid((3, 3),  # 将窗口分为3行3列 
                       (0, 0),   # 从第0行第0列开始作图
                       colspan=3)  # 列跨度为3
ax1.plot([1, 2], [1, 2])  
ax1.set_title('ax1_title')  # 设置标题
ax2 = plt.subplot2grid((3, 3), (1, 0), colspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)
ax4 = plt.subplot2grid((3, 3), (2, 0))
ax5 = plt.subplot2grid((3, 3), (2, 1))
ax4.scatter([1, 2], [2, 2])
ax4.set_xlabel('ax4_x')
ax4.set_ylabel('ax4_y')
plt.show()

# gridspec
import matplotlib.gridspec as gridspec
plt.figure()
gs = gridspec.GridSpec(3, 3)  # 将窗口分为3行3列
ax6 = plt.subplot(gs[0, :])  # 创建一个图像，占第0行和所有列
ax7 = plt.subplot(gs[1, :2])  # 创建一个图像，占第1行和第2列前所有列
ax8 = plt.subplot(gs[1:, 2])  # 创建一个图像，占第1行后所有行和第2列
ax9 = plt.subplot(gs[-1, 0])  # 创建一个图像，占倒数第1行和第0列
ax10 = plt.subplot(gs[-1, -2])  # 创建一个图像，占倒数第1行和倒数第2列
plt.show()

# subplots
f, ((ax11, ax12), (ax13, ax14)) = plt.subplots(2, 2,   # 将窗口分为2行2列
                                               sharex=True,  # 共享x轴坐标
                                               sharey=True)  # 共享y轴坐标
ax11.scatter([1, 2], [1, 2])  # 创建一个散点图
plt.tight_layout()  # 紧凑显示图像(默认为松散)
plt.show()
```

```python
# Animation动画
from matplotlib import animation
fig, ax = plt.subplots()
x = np.arange(0, 2 * np.pi, 0.01)
line, = ax.plot(x, np.sin(x))

# 动画函数，更新每一帧上x对应的y值
def animate(i):
    line.set_ydata(np.sin(x + i / 10.0))
    return line,

# 开始帧函数
def init():
    line.set_ydata(np.sin(x))
    return line,

ani = animation.FuncAnimation(fig=fig,  # 进行动画绘制的figure
                              func=animate,  # 自定义动画函数
                              frames=100,  # 动画长度，一次循环包含的帧数
                              init_func=init,  # 自定义开始帧
                              interval=20,  # 更新频率
                              blit=False)  # 更新所有点还是仅更新变化的点（True）
plt.show()
ani.save('basic_animation.mp4', fps=30, extra_args=['-vcodec', 'libx264'])
# PyCharm设置 Perference->Tools->Python Scientific
# ->取消勾选 show plots in toolwindow
```
